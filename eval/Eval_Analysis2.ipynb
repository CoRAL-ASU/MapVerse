```python
def normalize_item(item):
    """Normalize an item by lowercasing and removing punctuation."""
    import string, re
    item = item.lower()
    item = ''.join(ch for ch in item if ch not in string.punctuation)
    item = re.sub(r'\b(a|an|the)\b', '', item)
    return item.strip()

def normalize_list(lst):
    """Normalize each element in the list."""
    return [normalize_item(x) for x in lst if x]

def compute_exact_match(pred_list, gold_list):
    return int(set(normalize_list(pred_list)) == set(normalize_list(gold_list)))

def compute_set_metrics(pred_list, gold_list):
    pred_list = [s for s in pred_list if s.strip()]
    gold_list = [s for s in gold_list if s.strip()]
    pred_set = set(normalize_list(pred_list))
    gold_set = set(normalize_list(gold_list))

    # print(pred_set)
    # print(gold_set)
    # print("--------------------------------------")

    intersection = len(pred_set & gold_set)
    union = len(pred_set | gold_set)

    jaccard = intersection / union if union > 0 else 0.0
    precision = intersection / len(pred_set) if len(pred_set) > 0 else 0.0
    recall = intersection / len(gold_set) if len(gold_set) > 0 else 0.0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0.0

    return {
        "Jaccard": jaccard,
        "Precision": precision,
        "Recall": recall,
        "F1": f1
    }

def evaluate_array_of_lists(predictions, ground_truths):
    """predictions and ground_truths are arrays of lists of items."""
    assert len(predictions) == len(ground_truths), "Predictions and ground truths length mismatch"

    total_em = total_jaccard = total_precision = total_recall = total_f1 = 0
    n = len(predictions)

    for pred, gold in zip(predictions, ground_truths):
        total_em += compute_exact_match(pred, gold)
        metrics = compute_set_metrics(pred, gold)
        total_jaccard += metrics["Jaccard"]
        total_precision += metrics["Precision"]
        total_recall += metrics["Recall"]
        total_f1 += metrics["F1"]

    return {
        "Exact Match": (total_em *100)/ n,
        "Jaccard": total_jaccard / n,
        "Precision": total_precision / n,
        "Recall": total_recall / n,
        "F1": total_f1 / n
    }
# lines.pop(3)
predictions = [str(x).split(",") for x in listing_df['llm_answer']]
ground_truths = [str(x).split(",") for x in listing_df['correct_answer']]
# # ---------- Example ----------
# predictions = [
#     ["Paris", "France"],          # prediction 1
#     ["China", "India"],           # prediction 2
#     ["Washington", "D.C."]        # prediction 3
# ]

# ground_truths = [
#     ["France", "Paris"],          # gold 1
#     ["India", "China"],           # gold 2
#     ["Washington D.C."]           # gold 3
# ]

metrics = evaluate_array_of_lists(predictions, ground_truths)
print(metrics)
```

```python
def ranking_correlations2(true_items, pred_items):
    true_list = to_list(true_items)
    pred_list = to_list(pred_items)
    pred_list1 = pred_list[:]
    pred_list2 = pred_list[:]
    if len(true_list) > len(pred_list):
      # Find elements in true_list that are not in pred_list
      missing_elements_set = set(true_list) - set(pred_list)

      # Filter the true_list to get the missing elements in their original order
      missing_elements = [item for item in true_list if item in missing_elements_set]

      # Reverse the order of the missing elements
      missing_elements_reversed = missing_elements[::-1]

      # Append the reversed elements to the end of pred_list
      pred_list1 += missing_elements_reversed
      pred_list2 += missing_elements

    # Convert to rank positions
    # print(true_list)
    # print(pred_list1)
    # print(pred_list2)
    # print("===============")
    true_ranks = get_rank_positions(true_list, true_list)
    pred_ranks1 = get_rank_positions(true_list, pred_list1)
    pred_ranks2 = get_rank_positions(true_list, pred_list2)
    # print(true_ranks)
    # print(pred_ranks1)
    # print(pred_ranks2)
    # print("===============")
    rho1, _ = spearmanr(true_ranks, pred_ranks1)
    tau1, _ = kendalltau(true_ranks, pred_ranks1)
    rho2, _ = spearmanr(true_ranks, pred_ranks2)
    tau2, _ = kendalltau(true_ranks, pred_ranks2)
    rho, tau = (rho1 + rho2) / 2, (tau1 + tau2) / 2
    return rho, tau
```

```python
from scipy.stats import spearmanr, kendalltau

def to_list(x):
    """Convert comma-separated string to list if needed."""
    if isinstance(x, str):
        return [item.strip() for item in x.split(",") if item.strip()]
    return list(x)

def get_rank_positions(reference_list, target_list):
    """Convert item ordering into rank positions based on reference order."""
    ref_index = {item: idx for idx, item in enumerate(reference_list)}
    # print("ref idx: ", ref_index)
    return [ref_index[item] for item in target_list if item in ref_index]

def precision_at_k(y_true, y_pred, k):
    y_true = set(to_list(y_true))
    y_pred_top_k = to_list(y_pred)[:k]
    relevant = sum(1 for item in y_pred_top_k if item in y_true)
    return relevant / k

def recall_at_k(y_true, y_pred, k):
    y_true = set(to_list(y_true))
    y_pred_top_k = to_list(y_pred)[:k]
    relevant = sum(1 for item in y_pred_top_k if item in y_true)
    return relevant / len(y_true) if len(y_true) > 0 else 0

def ranking_correlations(true_items, pred_items):
    true_list = to_list(true_items)
    pred_list = to_list(pred_items)
    # Convert to rank positions
    # print(true_list)
    # print(pred_list)
    # print("===============")
    true_ranks = get_rank_positions(true_list, true_list)
    pred_ranks = get_rank_positions(true_list, pred_list)
    # print(true_ranks)
    # print(pred_ranks)
    # print("===============")
    rho, _ = spearmanr(true_ranks, pred_ranks)
    tau, _ = kendalltau(true_ranks, pred_ranks)
    return rho, tau

# ---------------------------
# Example usage
# ---------------------------

true_items_str = "A,B,C,D"
pred_items_str = "A,C,B,D"

spearman, kendall = ranking_correlations2(true_items_str, pred_items_str)
p_at_2 = precision_at_k("A,B", pred_items_str, k=2)
r_at_2 = recall_at_k("A,B", pred_items_str, k=2)

print(f"Spearman's rho: {spearman:.3f}")
print(f"Kendall's tau: {kendall:.3f}")
print(f"Precision@2: {p_at_2:.3f}")
print(f"Recall@2: {r_at_2:.3f}")
```

```python
def mean_precision(pred, true):
    pred = list(pred.split(','))
    true = list(true.split(','))  # for faster lookup
    # print(pred)
    # print(true)
    max_len = max(len(pred), len(true))

    precisions = []
    for k in range(1, max_len + 1):
        retrieved = pred[:k]
        rank_true = true[:k]
        num_relevant = sum(item in rank_true for item in retrieved)
        # print(num_relevant, k)
        precisions.append(num_relevant / k)

    return sum(precisions) / len(precisions)
```

```python

def ask_gemini_vqa(model, prompt, answer1, answer2):
    """
    Performs Visual Question Answering using the Gemini Pro Vision model.

    Args:
        model: The configured GenerativeModel instance.
        question (str): The text question to ask about the image.
        image_path (str): The file path to the local image.

    Returns:
        str: The generated text answer from the model, or an error message.
    """
    print(f"-> Asking: {answer1}, {answer2}\n")

    # --- Prepare the Prompt ---
    # The Gemini Pro Vision model expects a list containing text and images.
    prompt_parts = prompt + "s1: "+str(answer1 )+ "\n" +"s2: " +str(answer2) + "\n"


    # --- Call the Model ---
    try:
        # The generate_content method can handle the list of parts directly
        response = model.generate_content(prompt_parts)
        # For nicely formatted output
        return response.text #textwrap.indent(response.text, '> ', predicate=lambda _: True)
    except Exception as e:
        return f"An error occurred while calling the Gemini API: {e}"
```
